```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
```

```python
qc = QuantumCircuit(7, 1)

d0, d1, d2, d3 = 0,1,2,3
s0, s1 = 4,5
y = 6

qc.x(d1)
qc.x(d2)
qc.x(s1)

qc.x(s0)
qc.x(s1)
qc.mcx([d0, s0, s1], y)
qc.x(s0)
qc.x(s1)

qc.x(s1)
qc.mcx([d1, s0, s1], y)
qc.x(s1)

qc.x(s0)
qc.mcx([d2, s1, s0], y)
qc.x(s0)

qc.mcx([d3, s0, s1], y)

qc.measure(y, 0)
```

```python
sim = Aer.get_backend('aer_simulator')

compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1).result()

counts = result.get_counts()
print("MUX Output (y):", counts)
```
